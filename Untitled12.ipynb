```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

results = {
    "Model": [],
    "Metric": [],
    "Score": []
}

for name, clf in models.items():
    y_pred = clf.predict(X_test)
    y_pred_prob = clf.predict_proba(X_test)[:, 1] if hasattr(clf, "predict_proba") else None

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    if multi_class:
        auc_score = roc_auc_score(y_test_bin, clf.predict_proba(X_test), average='weighted', multi_class='ovr') if hasattr(clf, "predict_proba") else None
    else:
        auc_score = roc_auc_score(y_test, y_pred_prob) if y_pred_prob is not None else None

    for metric, score in zip(["Accuracy", "Precision", "Recall", "F1-score", "AUC"],
                              [accuracy, precision, recall, f1, auc_score]):
        if score is not None:
            results["Model"].append(name)
            results["Metric"].append(metric)
            results["Score"].append(score)

results_df = pd.DataFrame(results)

plt.figure(figsize=(12, 6))
sns.barplot(data=results_df, x="Model", y="Score", hue="Metric", palette="viridis")

plt.xticks(rotation=45, ha="right")
plt.xlabel("Machine Learning Models")
plt.ylabel("Score")
plt.title("Evaluation Metrics Comparison for Different Models")
plt.legend(title="Metric")
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
